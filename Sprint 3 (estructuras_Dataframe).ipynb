### b) Datos faltantes por columna


```python
vuelos_colum_relev.isna().sum()
```

```python
print('Si sumamos todos los datos faltantes, hay un total de', vuelos_colum_relev.isna().sum().sum(), 'valores carentes de contenido.')
```

Como la cantidad de datos nulos son pocos en proporción a la cantidad de datos totales, me inclino más por la aplicación de la función bfill(), mediante el cual se aplica un valor dentro de un rango coherente al tomarlo del valor de la fila siguiente dentro de su respectiva columna. Así, por ejemplo, se evita eliminar toda la fila con la función dropna(), o bien asignar un valor generalizado para todos los datos con el valor nulo (independientemente de la columna a la que pertenezca) con la función fillna().

```python
vuelos_colum_relev=vuelos_colum_relev.bfill()
print('Ahora ya no existen datos faltantes')
vuelos_colum_relev.isna().sum()
```

### c) Creación de nuevas columnas

```python
# Velocidad media por vuelo
vuelos_colum_relev['AvarageSpeed']= (vuelos_colum_relev['Distance'] / vuelos_colum_relev['AirTime']*60).round(2)
vuelos_colum_relev
```

```python
# Vuelo con retraso o dentro de hora, entendiendo que la hora de llegada efectiva (ArrTime) es la hora real de llegada
# con o sin retraso

import numpy as np

vuelos_colum_relev['Delay_WithinTime']=np.where(vuelos_colum_relev['ArrTime']>vuelos_colum_relev['CRSArrTime'], 'Delay', 'WithinTime')
vuelos_colum_relev
```

```python
# Tiempo de carreteo o movimiento del avión sobre tierra
vuelos_colum_relev['TaxiIn_and_TaxiOut']=(vuelos_colum_relev['ActualElapsedTime'] - vuelos_colum_relev['AirTime']).round(2)
vuelos_colum_relev
```

### f) Nuevos datos que considere relevantes

#### Cantidad de vuelos por mes

```python
vuelos_mes = vuelos['Month'].value_counts().reset_index()
vuelos_mes.columns=['Número del mes', 'Cantidad de vuelos']
vuelos_mes.sort_values('Cantidad de vuelos', ascending=False)

#Los meses de Diciembre, Junio y Marzo son en los que más vuelos se realizan, mientras que en el mes de Septiembre es el que menos.
```

#### Cantidad vuelos por día del mes

```python

vuelos_Dia_Mes=vuelos['DayofMonth'].value_counts().reset_index()
vuelos_Dia_Mes.columns=['Día del Mes', 'Cantidad de vuelos']
vuelos_Dia_Mes.sort_values('Cantidad de vuelos', ascending=False)

#Llama la atención como precisamente los últimos días de cada mes (31, 30 y 29) es cuando se realizan menos vuelos,
#en cambio el día 1 está entre los tres primeros
```

#### Cantidad vuelos por día de la semana

```python
vuelos_Dia_Sem=vuelos['DayOfWeek'].value_counts().reset_index()
vuelos_Dia_Sem.columns=['Día de la semana', 'Cantidad de vuelos']
vuelos_Dia_Sem.sort_values('Cantidad de vuelos', ascending=False)

## En cuanto a los vuelos que se efectúan por día de la semana, el primer y último día laborables de la semana son los días
# que más vuelos se ejecutan. Por su parte, el sábado es el que menos.
```
