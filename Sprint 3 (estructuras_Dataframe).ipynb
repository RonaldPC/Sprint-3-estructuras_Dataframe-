### c) CreaciÃ³n de nuevas columnas

```python
vuelos['AvarageSpeed']= (vuelos['Distance'] / vuelos['AirTime']*60).round(2)
vuelos
```

### d) Tabla de aerolÃ­neas con mÃ¡s retrasos acumulados. Muestra de mayor a menor

```python
#CreaciÃ³n de un dataframe solo con valores 'Delay' (retraso) de la columna ya creada anteriormente 'Delay_WithinTime'

solo_retrasos=vuelos.groupby('Delay_WithinTime').get_group('Delay')
solo_retrasos
```

```python
# CreaciÃ³n de un dataframe solo con valores 'Delay' (retraso) de la columna ya creada anteriormente 'Delay_WithinTime'
solo_retrasos=vuelos.groupby('Delay_WithinTime').get_group('Delay')


# ContabilizaciÃ³n vuelos con retrasos por aerolÃ­nea
retrasos_aerol=pd.DataFrame(solo_retrasos.groupby('UniqueCarrier')['Delay_WithinTime'].value_counts())


# ContabilizaciÃ³n vuelos totales por aerolÃ­nea
total_vuelos_aerol=pd.DataFrame(vuelos.groupby('UniqueCarrier')['Delay_WithinTime'].count())

# Porcentaje de vuelos retrasados con respecto a los vuelos totales por aerolÃ­nea
porcent_retr=(retrasos_aerol/total_vuelos_aerol)*100

porcent_retr.columns=['Porcentaje']
porcent_retr.sort_values(by='Porcentaje', ascending=False)

```

### f) Nuevos datos que considere relevantes

#### Cantidad de vuelos por mes

```python
vuelos_mes = vuelos['Month'].value_counts().reset_index()
vuelos_mes.columns=['NÃºmero del mes', 'Cantidad de vuelos']
vuelos_mes.sort_values('Cantidad de vuelos', ascending=False)
```

Los meses de Diciembre, Junio y Marzo son en los que mÃ¡s vuelos se realizan, mientras que en el mes de Septiembre es el que menos.

#### Cantidad vuelos por dÃ­a del mes

```python

vuelos_Dia_Mes=vuelos['DayofMonth'].value_counts().reset_index()
vuelos_Dia_Mes.columns=['DÃ­a del Mes', 'Cantidad de vuelos']
vuelos_Dia_Mes.sort_values('Cantidad de vuelos', ascending=False)
```

Llama la atenciÃ³n como precisamente los Ãºltimos dÃ­as de cada mes (31, 30 y 29) es cuando se realizan menos vuelos, en cambio el dÃ­a 1 estÃ¡ entre los tres primeros.
